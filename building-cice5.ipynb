# Building CICE5

This should be more chill compared to MOM5 as the build instructions are clearer. https://github.com/ACCESS-NRI/cice5/blob/master/Makefile

````bash
cd /scratch/tm70/$USER
git clone https://github.com/ACCESS-NRI/cice5.git --recursive
cd cice5
export LIBACCESSOM2_ROOT=/scratch/tm70/$USER/libaccessom2
make 1deg
````

seems like parallelio should be being built, since there's a block that looks like:
```
if ( $IO_TYPE == 'pio' ) then
    # Build PIO
    mkdir -p ParallelIO/build
    cd ParallelIO/build
    setenv CC mpicc
    setenv FC mpifort
    cmake -DWITH_PNETCDF=OFF \
          -DPIO_ENABLE_TIMING=OFF \
          -DNetCDF_C_LIBRARY="${NETCDF}/lib/ompi3/libnetcdf.so" \
          -DNetCDF_C_INCLUDE_DIR="${NETCDF}/include/" \
          -DNetCDF_Fortran_LIBRARY="${NETCDF}/lib/ompi3/Intel/libnetcdff.so" \
          -DNetCDF_Fortran_INCLUDE_DIR="${NETCDF}/include/Intel" \
          -DCMAKE_INSTALL_PREFIX="${SRCDIR}/ParallelIO/build" ../
    make && make install
    cd -
endif
```
but the `ParallelIO` directory is empty!

````bash
# figure out how to get source code from spack
. /g/data/tm70/$USER/spack/0.22/spack-config/spack-enable.bash
# check which version is installed
spack find parallelio
# find source code url
spack info parallelio
````

````bash
# download source and untar
wget https://github.com/NCAR/ParallelIO/archive/pio2_5_2.tar.gz -O - | tar xzvf -
````

````bash
rm -rf ParallelIO
mv ParallelIO-pio2_5_2 ParallelIO
````

````bash
# let's try build again
make 1deg
````

That seeemed to work!
